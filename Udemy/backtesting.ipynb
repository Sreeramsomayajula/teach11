```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
data = yf.download('GRSE.NS',start='2020-01-01',end='2024-11-27')
df = pd.DataFrame(data)
df
```

```python
df.head()
df['100sma']=df['Close'].rolling(window=100).mean()
df['50sma']=df['Close'].rolling(window=50).mean()
df['21sma']=df['Close'].rolling(window=21).mean()
```

```python
df['signal']=np.where(df['50sma']>df['100sma'],1,0)
df['position']=df['signal'].diff()
df.to_csv('sma.csv')
```

```python
df['buyprice']=np.where(df['position']==1,df['Close'],0 )
df['sellprice']=np.where(df['position']==-1, df['Close'], 0)
```
